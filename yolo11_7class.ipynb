```python
import os
import csv
from ultralytics import YOLO

# ğŸ“Œ Google Drive iÃ§indeki eÄŸitim sonuÃ§larÄ±nÄ±n kaydedileceÄŸi klasÃ¶r
drive_dir = "/content/drive/MyDrive/yolov11/"
csv_file = os.path.join(drive_dir, "training_metrics_11x.csv")

# ğŸ“Œ CSV dosyasÄ±nÄ± oluÅŸtur (baÅŸlÄ±k ekleyerek)
if not os.path.exists(csv_file):
    with open(csv_file, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["epoch", "train_loss", "val_loss", "mAP50", "precision", "recall"])

# âœ… Modeli yÃ¼kle
model = YOLO(model_path)

# âœ… EÄŸitimi Ã§alÄ±ÅŸtÄ±r
results = model.train(
    data="/content/drive/MyDrive/7_class_dataset/yolo.yaml",
    epochs=1000,
    imgsz=1280,
    batch=8,
    mosaic=1.0,  # Mosaic augmentation aktif
    mixup=0.2,
    save=True,
    project=drive_dir,
    name="yolo11x_experiment"
)

# âœ… EÄŸitim sonrasÄ± metrikleri al
metrics = results.metrics  # YOLOv8'in yeni formatÄ±na uygun ÅŸekilde metrikleri al

# ğŸ“Œ EÄŸitim metriklerini kaydet
with open(csv_file, mode="a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow([
        results.epochs,  # Toplam epoch sayÄ±sÄ±
        metrics.box.loss,  # Kutu kaybÄ±
        metrics.cls.loss,  # SÄ±nÄ±f kaybÄ±
        metrics.map50,  # mAP@50
        metrics.precision,  # Precision
        metrics.recall  # Recall
    ])

print("âœ… EÄŸitim tamamlandÄ±! Metrikler baÅŸarÄ±yla kaydedildi.")
```

```python
import os
from ultralytics import YOLO

# Google Drive iÃ§indeki model ve test verileri
best_model_path = "/content/drive/MyDrive/yolov11/yolo11x_experiment2/weights/best.pt"
test_images_path = "/content/drive/MyDrive/7_class_dataset/images/test/"
result_dir = "/content/drive/MyDrive/yolov11_result/"

# ğŸ“Œ Ã‡Ä±ktÄ±larÄ±n kaydedileceÄŸi klasÃ¶r (Annotation TXT dosyalarÄ± burada olacak)
annotations_dir = os.path.join(result_dir, "test_annotations_1")
os.makedirs(annotations_dir, exist_ok=True)

# SonuÃ§ klasÃ¶rÃ¼nÃ¼ oluÅŸtur
os.makedirs(result_dir, exist_ok=True)

# EÄŸitilmiÅŸ modeli yÃ¼kle
model = YOLO(best_model_path)

# âœ… Test veri setindeki tÃ¼m gÃ¶rselleri al
test_images = sorted(os.listdir(test_images_path))
print(f"ğŸ“‚ Test klasÃ¶rÃ¼nde {len(test_images)} adet dosya var.")
# Test veri setinde tahmin yap ve sonuÃ§larÄ± kaydet
results=model.predict(
    source=test_images_path,
    conf=0.10,  # GÃ¼ven skoru eÅŸiÄŸi (0.25 Ã¼zeri tahminler gÃ¶sterilir)
    save=True,  # SonuÃ§larÄ± kaydet
    project=result_dir,
    name="test_results_11x-7class-noname",
    show_labels=False,  # **Etiket isimlerini gizle**
    show_conf=False  # **OlasÄ±lÄ±k deÄŸerlerini gizle**
)
# âœ… Tahmin sonuÃ§larÄ±nÄ± YOLO formatÄ±nda kaydet
# âœ… Tahmin sonuÃ§larÄ±nÄ± YOLO formatÄ±nda kaydet

# âœ… Annotation'larÄ± kaydet
processed_files = set()
for result in results:
    image_path = result.path  # GÃ¶rÃ¼ntÃ¼ dosyasÄ± yolu
    image_name = os.path.basename(image_path).split(".")[0]  # Dosya adÄ± (uzantÄ±sÄ±z)

    # ğŸ“Œ Resim boyutlarÄ±nÄ± al
    img_width, img_height = result.orig_shape  # YOLO Ã§Ä±ktÄ±larÄ±ndan orijinal gÃ¶rÃ¼ntÃ¼ boyutu

    # ğŸ“Œ Annotation dosya yolu (TXT olarak kaydedilecek)
    txt_path = os.path.join(annotations_dir, f"{image_name}.txt")

    # âœ… Ä°ÅŸlenen dosyalarÄ± takip edelim
    processed_files.add(image_name)

    with open(txt_path, "w") as f:
        if result.boxes is None or len(result.boxes) == 0:
            print(f"âš ï¸ {image_name} iÃ§in tahmin yok, boÅŸ dosya oluÅŸturuldu.")
            continue  # HiÃ§ nesne tespit edilmemiÅŸse boÅŸ bÄ±rak

        for box in result.boxes.data:
            x_min, y_min, x_max, y_max, conf, class_id = box.tolist()

            # ğŸ”„ KoordinatlarÄ± YOLO formatÄ±na dÃ¶nÃ¼ÅŸtÃ¼r (Normalize edilmiÅŸ deÄŸerler)
            x_center = ((x_min + x_max) / 2) / img_width
            y_center = ((y_min + y_max) / 2) / img_height
            width = (x_max - x_min) / img_width
            height = (y_max - y_min) / img_height

            # ğŸ“Œ SÄ±nÄ±f ID'sini tamsayÄ±ya Ã§evir ve dosyaya kaydet
            f.write(f"{int(class_id)} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")

print(f"âœ… TÃ¼m test annotationlarÄ± YOLO formatÄ±nda kaydedildi: {annotations_dir}")
```
