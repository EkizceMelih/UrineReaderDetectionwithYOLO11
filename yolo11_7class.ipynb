```python
import os
import csv
from ultralytics import YOLO

# 📌 Google Drive içindeki eğitim sonuçlarının kaydedileceği klasör
drive_dir = "/content/drive/MyDrive/yolov11/"
csv_file = os.path.join(drive_dir, "training_metrics_11x.csv")

# 📌 CSV dosyasını oluştur (başlık ekleyerek)
if not os.path.exists(csv_file):
    with open(csv_file, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["epoch", "train_loss", "val_loss", "mAP50", "precision", "recall"])

# ✅ Modeli yükle
model = YOLO(model_path)

# ✅ Eğitimi çalıştır
results = model.train(
    data="/content/drive/MyDrive/7_class_dataset/yolo.yaml",
    epochs=1000,
    imgsz=1280,
    batch=8,
    mosaic=1.0,  # Mosaic augmentation aktif
    mixup=0.2,
    save=True,
    project=drive_dir,
    name="yolo11x_experiment"
)

# ✅ Eğitim sonrası metrikleri al
metrics = results.metrics  # YOLOv8'in yeni formatına uygun şekilde metrikleri al

# 📌 Eğitim metriklerini kaydet
with open(csv_file, mode="a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow([
        results.epochs,  # Toplam epoch sayısı
        metrics.box.loss,  # Kutu kaybı
        metrics.cls.loss,  # Sınıf kaybı
        metrics.map50,  # mAP@50
        metrics.precision,  # Precision
        metrics.recall  # Recall
    ])

print("✅ Eğitim tamamlandı! Metrikler başarıyla kaydedildi.")
```

```python
import os
from ultralytics import YOLO

# Google Drive içindeki model ve test verileri
best_model_path = "/content/drive/MyDrive/yolov11/yolo11x_experiment2/weights/best.pt"
test_images_path = "/content/drive/MyDrive/7_class_dataset/images/test/"
result_dir = "/content/drive/MyDrive/yolov11_result/"

# 📌 Çıktıların kaydedileceği klasör (Annotation TXT dosyaları burada olacak)
annotations_dir = os.path.join(result_dir, "test_annotations_1")
os.makedirs(annotations_dir, exist_ok=True)

# Sonuç klasörünü oluştur
os.makedirs(result_dir, exist_ok=True)

# Eğitilmiş modeli yükle
model = YOLO(best_model_path)

# ✅ Test veri setindeki tüm görselleri al
test_images = sorted(os.listdir(test_images_path))
print(f"📂 Test klasöründe {len(test_images)} adet dosya var.")
# Test veri setinde tahmin yap ve sonuçları kaydet
results=model.predict(
    source=test_images_path,
    conf=0.10,  # Güven skoru eşiği (0.25 üzeri tahminler gösterilir)
    save=True,  # Sonuçları kaydet
    project=result_dir,
    name="test_results_11x-7class-noname",
    show_labels=False,  # **Etiket isimlerini gizle**
    show_conf=False  # **Olasılık değerlerini gizle**
)
# ✅ Tahmin sonuçlarını YOLO formatında kaydet
# ✅ Tahmin sonuçlarını YOLO formatında kaydet

# ✅ Annotation'ları kaydet
processed_files = set()
for result in results:
    image_path = result.path  # Görüntü dosyası yolu
    image_name = os.path.basename(image_path).split(".")[0]  # Dosya adı (uzantısız)

    # 📌 Resim boyutlarını al
    img_width, img_height = result.orig_shape  # YOLO çıktılarından orijinal görüntü boyutu

    # 📌 Annotation dosya yolu (TXT olarak kaydedilecek)
    txt_path = os.path.join(annotations_dir, f"{image_name}.txt")

    # ✅ İşlenen dosyaları takip edelim
    processed_files.add(image_name)

    with open(txt_path, "w") as f:
        if result.boxes is None or len(result.boxes) == 0:
            print(f"⚠️ {image_name} için tahmin yok, boş dosya oluşturuldu.")
            continue  # Hiç nesne tespit edilmemişse boş bırak

        for box in result.boxes.data:
            x_min, y_min, x_max, y_max, conf, class_id = box.tolist()

            # 🔄 Koordinatları YOLO formatına dönüştür (Normalize edilmiş değerler)
            x_center = ((x_min + x_max) / 2) / img_width
            y_center = ((y_min + y_max) / 2) / img_height
            width = (x_max - x_min) / img_width
            height = (y_max - y_min) / img_height

            # 📌 Sınıf ID'sini tamsayıya çevir ve dosyaya kaydet
            f.write(f"{int(class_id)} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")

print(f"✅ Tüm test annotationları YOLO formatında kaydedildi: {annotations_dir}")
```
